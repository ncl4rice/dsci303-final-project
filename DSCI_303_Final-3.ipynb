```python
# Start by classifying the start state and end state of each event ----

import numpy as np
import pandas as pd

data = (
    events
        #.sort_values('index')
        .assign(
            # Extract x and y locations as numerics
            # It's a little bit tricky because some of the entries are NaN, not list
            loc_x=lambda x: x['location'].apply(lambda y: y[0] if isinstance(y, list) else np.nan),
            loc_y=lambda x: x['location'].apply(lambda y: y[1] if isinstance(y, list) else np.nan),
            # Unlike Rudd's work, let's split the field into 30 zones (6 for x, 5 for y)
            zone_x=lambda x: np.select(
                condlist=[x['loc_x'] < 20, x['loc_x'] < 40, x['loc_x'] < 60, x['loc_x'] < 80, x['loc_x'] < 100, x['loc_x'] < 120],
                choicelist=['0', '1', '2', '3', '4', '5'],
                default=None
            ),
            zone_y=lambda x: np.select(
                condlist=[x['loc_y'] < 16, x['loc_y'] < 32, x['loc_y'] < 48, x['loc_y'] < 64, x['loc_y'] < 80],
                choicelist=['0', '1', '2', '3', '4'],
                default=None
            ),
            zone=lambda x: x['zone_x'] + x['zone_y'],
            recipient_player_id = lambda x: np.select(
            condlist=[
                (x['type'] == 'Pass') & (x['team_id'] == x['team_id'].shift(1)) & (x['match_id'] == x['match_id'].shift(1))
            ],
            choicelist=[x['player_id'].shift(1)],
            default=x['player_id']
            ),
            distance_from_own_goal=lambda x: np.sqrt((x['loc_x'] - 0)**2 + (x['loc_y'] - 40)**2),
            pass_distance=lambda x: np.sqrt((x['loc_x'] - x['loc_x'].shift(1))**2 + (x['loc_y'] - x['loc_y'].shift(-1))**2),
            recipient_player = lambda x: np.select(
                condlist=[(x['type'] == 'Pass') & (x['team_id'] == x['team_id'].shift(1)) & (x['match_id'] == x['match_id'].shift(1))],
                choicelist=[x['player'].shift(1)],
                default= x['player'] ),
            # half = lambda x: np.select(
            #     condlist = [pd.to_datetime(x['match_date']) <= '12/31/15',  pd.to_datetime(x['match_date']) > '12/31/15'],
            #     choicelist = ['First Half', 'Second Half'],
            #     default = None
            # )
            # passer_player_id = lambda x: np.select(
            # condlist=[
            #     (x['type'] == 'Pass') & (x['team_id'] == x['team_id'].shift(-1)) & (x['match_id'] == x['match_id'].shift(-1))
            # ],
            # choicelist=[x['player_id'].shift(-1)],
            # default=x['player_id']
            # ),
            # passer_player = lambda x: np.select(
            #     condlist=[(x['type'] == 'Pass') & (x['team_id'] == x['team_id'].shift(-1)) & (x['match_id'] == x['match_id'].shift(-1))],
            #     choicelist=[x['player'].shift(-1)],
            #     default= x['player'] )
        )
        # According to Rudd's Markov model, we only consider these three event types
        .loc[lambda x: x['type'].isin(['Pass', 'Carry', 'Shot'])]
        .assign(
            # The starting state is the zone on the field where the event started, unless it's
            # one of the special game states (Free Kick, Corner, Throw-in). We skip Penalty because
            # there were no penalties in this game, and for simplicity we do not distinguish
            # between short and long like Rudd did.
            end_loc_x = lambda x: x['loc_x'].shift(-1),
            end_loc_y = lambda x: x['loc_y'].shift(-1),
            start_state=lambda x: np.select(
                condlist=[x['pass_type'] == 'Free Kick', x['pass_type'] == 'Corner', x['pass_type'] == 'Throw-in'],
                choicelist=['Free Kick', 'Corner', 'Throw-in'],
                default=x['zone']
            ),
            # We define the ending state to be the starting state of the next event. We make
            # exceptions for terminal states, which happen on a goal or a change of possession.
            end_state=lambda x: np.select(
                condlist=[x['shot_outcome'] == 'Goal', ~(x['possession_team_id'] == x['possession_team_id'].shift(-1))],
                choicelist=['Goal', 'End'],
                default=x['start_state'].shift(-1)
            ),
            success = lambda x: np.select(
                condlist=[(x['possession_team_id'] == x['possession_team_id'].shift(-1)), ~(x['possession_team_id'] == x['possession_team_id'].shift(-1))],
                choicelist=[1, 0],
                default = None
            )
        )
)
```

```python

import numpy as np

events = (
    events
        .sort_values(['match_id', 'index'])
        .assign(
            start_loc_x=lambda x: x['location'].apply(lambda y: y[0] if isinstance(y, list) else np.nan),
            start_loc_y=lambda x: x['location'].apply(lambda y: y[1] if isinstance(y, list) else np.nan),
            end_location=lambda x: np.select(
                condlist=[x['type'] == 'Pass', x['type'] == 'Carry', x['type'] == 'Shot', x['type'] == 'Goalkeeper'],
                choicelist=[x['pass_end_location'], x['carry_end_location'], x['shot_end_location'], x['goalkeeper_end_location']],
                default=x['location'],
            ),
            end_loc_x=lambda x: x['end_location'].apply(lambda y: y[0] if isinstance(y, list) else np.nan),
            end_loc_y=lambda x: x['end_location'].apply(lambda y: y[1] if isinstance(y, list) else np.nan),

            recipient_player_id = lambda x: np.select(
            condlist=[
                (x['type'] == 'Pass') & (x['team_id'] == x['team_id'].shift(1)) & (x['match_id'] == x['match_id'].shift(1))
            ],
            choicelist=[x['player_id'].shift(1)],
            default=x['player_id']
            ),
            recipient_player = lambda x: np.select(
                condlist=[(x['type'] == 'Pass') & (x['team_id'] == x['team_id'].shift(1)) & (x['match_id'] == x['match_id'].shift(1))],
                choicelist=[x['player'].shift(1)],
                default= x['player'] ),

            result=lambda x: np.select(
                condlist=[
                    x['type'] == 'Pass',
                    x['type'] == 'Carry',
                    x['type'] == 'Ball Recovery',
                    x['type'] == 'Duel',
                    x['type'] == 'Clearance',
                    x['type'] == 'Dribble',
                    x['type'] == 'Goalkeeper',
                    x['type'] == 'Dispossessed',
                    x['type'] == 'Miscontrol',
                    x['type'] == 'Shot',
                    x['type'] == 'Foul Committed',
                    x['type'] == 'Foul Won',
                    x['type'] == 'Dribbled Past',
                    x['type'] == 'Interception',
                ],
                choicelist=[
                    # Pass
                    np.select(
                        condlist=[x['pass_outcome'].isna(), x['pass_outcome'] == 'Incomplete', x['pass_outcome'] == 'Pass Offside'],
                        choicelist=['success', 'failure', 'offside'],
                        default=None
                    ),
                    'success',
                    'success',
                    np.select(
                        condlist=[x['duel_type'] == 'Aerial Lost', x['duel_outcome'].isin(['Lost In Play', 'Lost Out']), x['duel_outcome'].isin(['Won', 'Success in Play'])],
                        choicelist=['failure', 'failure', 'success'],
                        default=None
                    ),
                    'success',
                    np.where(x['dribble_outcome'] == 'Complete', 'success', 'failure'), # Dribble

                    np.select(
                        condlist=[x['goalkeeper_type'].isin(['Shot Faced', 'Shot Saved', 'Collected']), x['goalkeeper_outcome'] == 'Claim'],
                        choicelist=['success', 'success'],
                        default=None
                    ),
                    'failure',
                    'failure',
                    np.where(x['shot_outcome'] == 'Goal', 'success', 'failure'),    # Shot
                    'failure',
                    'success',
                    'success',
                    np.where(x['interception_outcome'] == 'Won', 'success', 'failure'), # Interception
                ],
                default=None
            ),
            type_result=lambda x: x['type'] + '_' + x['result'],
        )
        .loc[lambda x: ~x['result'].isna()]
)

```

```python

k = 10

events = (
    events
        .assign(

            start_loc_x_2=lambda x: x['start_loc_x'].shift(2),
            start_loc_y_2=lambda x: x['start_loc_y'].shift(2),

        ).assign(
            index_spadl=lambda x: range(x.shape[0]),
            goal_index_spadl=lambda x: np.where(x['shot_outcome'] == 'Goal', x['index_spadl'], np.nan),
            goal_match_id=lambda x: np.where(x['shot_outcome'] == 'Goal', x['match_id'], np.nan),
            goal_team_id=lambda x: np.where(x['shot_outcome'] == 'Goal', x['team_id'], np.nan),
            next_goal_index_spadl=lambda x: x['goal_index_spadl'].fillna(method='bfill'),
            next_goal_match_id=lambda x: x['goal_match_id'].fillna(method='bfill'),
            next_goal_team_id=lambda x: x['goal_team_id'].fillna(method='bfill'),
            is_goal_scored=lambda x: list(
                map(
                    int,
                    (x['next_goal_match_id'] == x['match_id']) &
                        (x['next_goal_team_id'] == x['team_id']) &
                        (x['next_goal_index_spadl'] <= x['index_spadl'] + 10)
                )
            ),
            is_goal_conceded=lambda x: list(
                map(
                    int,
                    (x['next_goal_match_id'] == x['match_id']) &
                        ~(x['next_goal_team_id'] == x['team_id']) &
                        (x['next_goal_index_spadl'] <= x['index_spadl'] + 10)

                )
            ),
        )
        .assign(
            end_distance_opp_goal=lambda x: np.sqrt((x['end_loc_x']-120)**2 + (x['end_loc_y'] - 40)**2),
            end_distance_own_goal=lambda x: np.sqrt((x['end_loc_x'])**2 + (x['end_loc_y'] - 40)**2),
            downfield_movement=lambda x: x['end_loc_x'] - x['start_loc_x_2'],
        )
        .loc[lambda x: ~x['downfield_movement'].isna()]
)
```

```python

import statsmodels.api as sm
import statsmodels.formula.api as smf

model_goal_scored = (
    smf.glm(
        formula="is_goal_scored ~ type_result + end_distance_opp_goal + downfield_movement",
        data=events.loc[lambda x: x['type_result'] != "Shot_success"],
        family=sm.families.Binomial()
    )
        .fit()
)

model_goal_conceded = (
    smf.glm(
        formula="is_goal_conceded ~ type_result + end_distance_own_goal + downfield_movement",
        data=events.loc[lambda x: x['type_result'] != "Shot_success"],
        family=sm.families.Binomial()
    )
        .fit()
)

model_goal_scored.summary()
```

```python

events_no_goal = (
    events
        .loc[lambda x: x['type_result'] != 'Shot_success']
        .assign(
            pred_goal_scored=model_goal_scored.fittedvalues,
            pred_goal_conceded=model_goal_conceded.fittedvalues,
        )
)

events_goal = (
    events
        .loc[lambda x: x['type_result'] == 'Shot_success']
        .assign(
            pred_goal_scored=1,
            pred_goal_conceded=events_no_goal.loc[lambda x: x['type'] == 'Shot']['pred_goal_conceded'].mean(),
        )
)


events = (
    pd.concat([events_goal, events_no_goal])
        .sort_values(['match_id', 'index'])
        .assign(

                diff_goal_scored=lambda x: np.select(
                condlist=[x['pass_type'] == 'Kick Off', x['team_id'] == x['team_id'].shift(1), ~(x['team_id'] == x['team_id'].shift(1))],
                choicelist=[0, x['pred_goal_scored'] - x['pred_goal_scored'].shift(1), x['pred_goal_scored'] - x['pred_goal_conceded']],
                default=None
            ),
            diff_goal_conceded=lambda x: np.select(
                condlist=[x['pass_type'] == 'Kick Off', x['team_id'] == x['team_id'].shift(1), ~(x['team_id'] == x['team_id'].shift(1))],
                choicelist=[0, x['pred_goal_conceded'] - x['pred_goal_conceded'].shift(1), x['pred_goal_conceded'] - x['pred_goal_scored']],
                default=None
            ),
            vaep=lambda x: x['diff_goal_scored'] - x['diff_goal_conceded']
        )
)
```

```python

import numpy as np

events = (
    events
        .sort_values(['match_id', 'index'])
        .assign(
            start_loc_x=lambda x: x['location'].apply(lambda y: y[0] if isinstance(y, list) else np.nan),
            start_loc_y=lambda x: x['location'].apply(lambda y: y[1] if isinstance(y, list) else np.nan),
            end_location=lambda x: np.select(
                condlist=[x['type'] == 'Pass', x['type'] == 'Carry', x['type'] == 'Shot', x['type'] == 'Goalkeeper'],
                choicelist=[x['pass_end_location'], x['carry_end_location'], x['shot_end_location'], x['goalkeeper_end_location']],
                default=x['location'],
            ),
            end_loc_x=lambda x: x['end_location'].apply(lambda y: y[0] if isinstance(y, list) else np.nan),
            end_loc_y=lambda x: x['end_location'].apply(lambda y: y[1] if isinstance(y, list) else np.nan),
            recipient_player_id = lambda x: np.select(
            condlist=[
                (x['type'] == 'Pass') & (x['team_id'] == x['team_id'].shift(1)) & (x['match_id'] == x['match_id'].shift(1))
            ],
            choicelist=[x['player_id'].shift(1)],
            default=x['player_id']
            ),
            recipient_player = lambda x: np.select(
                condlist=[(x['type'] == 'Pass') & (x['team_id'] == x['team_id'].shift(1)) & (x['match_id'] == x['match_id'].shift(1))],
                choicelist=[x['player'].shift(1)],
                default= x['player'] ),

            result=lambda x: np.select(
                condlist=[
                    x['type'] == 'Pass',
                    x['type'] == 'Carry',
                    x['type'] == 'Ball Recovery',
                    x['type'] == 'Duel',
                    x['type'] == 'Clearance',
                    x['type'] == 'Dribble',
                    x['type'] == 'Goalkeeper',
                    x['type'] == 'Dispossessed',
                    x['type'] == 'Miscontrol',
                    x['type'] == 'Shot',
                    x['type'] == 'Foul Committed',
                    x['type'] == 'Foul Won',
                    x['type'] == 'Dribbled Past',
                    x['type'] == 'Interception',
                ],
                choicelist=[
                    np.select(
                        condlist=[x['pass_outcome'].isna(), x['pass_outcome'] == 'Incomplete', x['pass_outcome'] == 'Pass Offside'],
                        choicelist=['success', 'failure', 'offside'],
                        default=None
                    ),
                    'success',
                    'success',

                    np.select(
                        condlist=[x['duel_type'] == 'Aerial Lost', x['duel_outcome'].isin(['Lost In Play', 'Lost Out']), x['duel_outcome'].isin(['Won', 'Success in Play'])],
                        choicelist=['failure', 'failure', 'success'],
                        default=None
                    ),
                    'success',
                    np.where(x['dribble_outcome'] == 'Complete', 'success', 'failure'),
                    np.select(
                        condlist=[x['goalkeeper_type'].isin(['Shot Faced', 'Shot Saved', 'Collected']), x['goalkeeper_outcome'] == 'Claim'],
                        choicelist=['success', 'success'],
                        default=None
                    ),
                    'failure',
                    'failure',
                    np.where(x['shot_outcome'] == 'Goal', 'success', 'failure'),
                    'failure',
                    'success',
                    'success',
                    np.where(x['interception_outcome'] == 'Won', 'success', 'failure'),
                ],
                default=None
            ),
            type_result=lambda x: x['type'] + '_' + x['result'],
        )
        .loc[lambda x: ~x['result'].isna()]
)

```
